```python
import pandas as pd
import os
```

```python
# 获取文件夹中的所有 CSV 文件
folder_path = 'D:/project/pythonProject1/汇总测试文件夹'
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
```

```python
# 创建一个空的 DataFrame 来存储汇总结果
combined_df = pd.DataFrame()
```

```python
for csv_file in csv_files:
    # 读取每个 CSV 文件
    df = pd.read_csv(os.path.join(folder_path, csv_file))
    # 添加期次列，去除.csv 后缀
    df['期次'] = csv_file.replace('.csv', '')
    # 拆分素材名称列
    df[['H5id', 'other']] = df['素材名称'].str.split('_', n=1, expand=True)
    # 根据实际情况再次拆分获取渠道名称
    df['other'].fillna('', inplace=True)
    df['other'] = df['other'].replace('', pd.NA)
    temp_df = df['other'].str.extract('^(.*?)(?:-\w+_\w+_\w+_\w+)?$', expand=False)
    df['渠道名称'] = temp_df
    # 删除临时列和不需要的列
    df.drop(['other', '投放表单回传数', '投放支付回传数', '投放加微回传数', '投放加微回传率'], axis=1, inplace=True)
    # 将 H5id 列转换为整数数组
    df['H5id'] = df['H5id'].astype(int)

    # 定义期望的列顺序
    desired_columns = ['期次', 'H5id', '渠道名称', '素材名称', '支付成功例子数', '有效例子数', '有效加微例子数',
                       '加微例子数', '加微率', '重复例子数(30天)', '历史例子数', '历史例子占比', '填写问卷数', 
                       '填写问卷率', '单向好友数', '导学课到课率', '导学课完课率', 'D1到课率', 'D1完课率', '正价转化数', 
                       '正价转化率', '正价转化金额', '正价课退单数', '退单金额', '退单率']

    # 如果当前 DataFrame 的列不在期望的列顺序中，添加缺失的列并填充为 NaN
    for col in desired_columns:
        if col not in df.columns:
            df[col] = pd.NA

    # 按照期望的列顺序重新排列列
    df = df[desired_columns]

    # 使用 pd.concat 来合并 DataFrame
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
# 保存汇总后的表格
combined_df.to_csv('汇总表格.csv', index=False)
```
